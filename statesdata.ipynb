```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load dataset
data = pd.read_csv('states.csv')
states_data = pd.DataFrame(data)
states_data
```

```python
#check data info
states_data.info()
```

```python
#determine the state with the highest health insurance coverage
highest = states_data.groupby('State')['Health Insurance Coverage Change (2010-2015)'].max()
highest
```

```python
#remove the united state row 
usa_row = states_data[states_data['State'] == 'United States' ]
```

```python
df = states_data[states_data['State'] != 'United States']
df
```

```python
#set index
state_index = df.set_index('State')
state_index
```

```python
#determine the state with the highest Health Insurance Coverage Change
highest_HICC = df['Health Insurance Coverage Change (2010-2015)'].max()
state_with_HICC = df[df['Health Insurance Coverage Change (2010-2015)']==highest_HICC]['State'].values[0]
print('state with the highest Health Insurance Coverage Change (2010-2015):',state_with_HICC)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#comparison of unisured rates(2010 and 2015) using line plot
plt.figure(figsize=(20, 25))

plt.plot(df['State'], df['Uninsured Rate (2010)'], marker='o', label='2010')
plt.plot(df['State'], df['Uninsured Rate (2015)'], marker='o', label='2015')

  
plt.xlabel('State')
plt.ylabel('Uninsured Rate')
plt.title('Comparison of Uninsured Rates (2010 vs 2015)')
plt.legend()
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
import seaborn as sns

# Box plot to compare uninsured rate change for Medicaid expansion states vs non-expansion states
plt.figure(figsize=(8, 8))
sns.boxplot(x=df['State Medicaid Expansion (2016)'], y=df['Uninsured Rate Change (2010-2015)'])
plt.xlabel('Medicaid Expansion')
plt.ylabel('Uninsured Rate Change')
plt.title('Impact of Medicaid Expansion on Uninsured Rate Change')
plt.show()
```

```python
# Stacked bar plot of employer coverage and marketplace coverage
plt.figure(figsize=(10, 6))
plt.bar(df['State'], df['Employer Health Insurance Coverage (2015)'], label='Employer Coverage')
plt.bar(df['State'], df['Marketplace Health Insurance Coverage (2016)'], label='Marketplace Coverage')
plt.xlabel('State')
plt.ylabel('Coverage Percentage')
plt.title('Employer Coverage vs. Marketplace Coverage')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot of tax credits vs. marketplace coverage
plt.figure(figsize=(8, 6))
plt.scatter(df['Marketplace Tax Credits (2016)'], df['Marketplace Health Insurance Coverage (2016)'])
plt.xlabel('Marketplace Tax Credits')
plt.ylabel('Marketplace Coverage')
plt.title('Tax Credits vs. Marketplace Coverage')
plt.show()
```

```python
# Scatter plot of Medicaid enrollment changes vs. uninsured rate changes
plt.figure(figsize=(8, 10))
plt.scatter(df['Medicaid Enrollment Change (2013-2016)'], df['Uninsured Rate Change (2010-2015)'])
plt.xlabel('Medicaid Enrollment Change')
plt.ylabel('Uninsured Rate Change')
plt.title('Medicaid Enrollment Changes vs. Uninsured Rate Changes')
plt.show()
```
